```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("F:/ML_project/project/seaborn/WorldCupMatches.csv")

df_goals = pd.read_csv("F:/ML_project/project/seaborn/goals.csv")

df["Total Goals"]=df['Home Team Goals'] + df['Away Team Goals']
print(df.head())
sns.set_style('whitegrid')
sns.set_context('poster', font_scale=0.4)

f , ax = plt.subplots(figsize=(12, 7))
ax = sns.barplot(data=df, x =df['Year'], y=df["Total Goals"])

ax.set_title('Average Goal Per year in WorldCup')

plt.show()



```

```python
f , ax2 = plt.subplots(figsize=(12, 7))
ax2 =sns.boxplot(data=df_goals, x= 'year', y = 'goals', palette ="Spectral")
ax2.set_title('Goals Visualization')
plt.show()
print(df_goals.head())
```
